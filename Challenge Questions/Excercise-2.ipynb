```
A login register is maintained in the library of VITCC in which, the
register number of students are recorded when they enter the library.
Sometimes it happens that the students visit the library more than
once in a day and hence duplicate entries occur so frequently in the
register. The librarian wants to have a report of all students who have
visited on a particular day, ‘x’. Given the list of students who visited
the library on the day ‘x’, write a python program to prepare a report
with unique register number of students. Also read a register number
‘r’ and search for it in the list. Print ‘Found’ if ‘r’ is in the list and print
‘Not found’ otherwise.
Input Format:
First line contains the number of students visited, ‘n’
Second line contains the register number of first entry
Third line contains the register number of second entry
. . .
N+1th line contains the register number of n-th entry
Next line contains the register number ‘r’ that has to be searched
Output Format:
A list without duplicate entries
Print either ‘Found’ or ‘Not found’
Boundary Conditions:
Number of students visited >=0
```

**Input:-**
```
n //number of students visited
lst //store reg number of students
ser //Number to search in List
```
**Logic:-**
```
st=set(lst)  //Converting list to set as set store only unique values
lst=list(st) //again rewriting the lst with unique value
 if ser in lst:              //check element to be searched in list 
    print("found")
  else:
    print("Not Found")
```

**Output:-**
```
lst //List of Entries(Unique value Only)
Found //if entry found 
not Found //if Entry not Found

```

````
'''
Test Cases Input
4
12BEC1121
12BME1015
12BME1005
12BCE1015
12BEC1121
Test Cases Expected Output
['12BEC1121', '12BME1015', '12BME1005',
'12BCE1015']
Found
'''
n=int(input("Enter number of students visited:-"))
lst=[]
if n>0:
  for i in range(0,n):
    lst.append(input("Enter Register number of student visited:-"))
  st=set(lst) 
  lst=list(st) 
  ser=input("Enter register number you want to search :-")
  print(lst)
  if ser in lst:
    print("found")
  else:
    print("Not Found")

````

```
Write a python code to count the number of
unique words in a given passage. The paragraph
may contain words with special characters such
as '!', '?', '.', ',' , ':' and ';' and digits are not

the end of a word in a passage that is Hello World!

No two special characters occur together. Print
'Invalid input' for such cases. Count words without

insensitive. Print words in lower case and in
sorted order. For example, given a passage
'Programming is a skill in demand! Your
programming skills are bankable assets.' The
output should be
{'a': 1,
'are': 1,
'assets': 1,
'bankable': 1,
'demand': 1,
'in': 1,

top of your program]

enter a paragraph
Output format:
Dictionary of words in the passage and their count
Boundary conditions:
One paragraph is entered at a strech

```

**Input:-**
```
para //TO store Para Entered bby User
dct //Dictionary to store words in pair where key is word and value is count 
```
**Logic:-**
```
for i in para.split():      //.split is used to take out words from string
  if i.startswith('!') or i.startswith('?') or i.startswith(':') or i.startswith(',') or i.startswith(';') :   //to check that word dont startswith Special char
    print("Invalid Output")
    flag=0    //To detect that there is invalid word in list
    break;
  else:  
    i=i.lower()     //to make word in one case

```

`dct //printing the dictionary with word as key and value as occurence of key   (word)`
