Given a string, draw the PAC, pseudocode and develop the
python code to check whether the given string is a magic string
or not. A string is said to be a magic string if the substrings
formed by characters in the even positions and odd positions are
the same. If the string is of odd length, the middle character has
to be skipped.
For example, the string ‘aabbcc’ is a magic string as the
substring ‘abc’ that is formed using character at even position
and the substring ‘abc’ that is formed using character at odd
position is same. However the string ‘abcabc’ is not a magic
string.


```
Input format:
Enter a string
Output format:
Magic string
Not a magic string
Test Cases Input
aabbcc
abcabc
aabcc
Test Cases Expected Output
Magic string
Not a magic string
Magic string
Test Cases Input aba
Test Cases Expected Output Not a magic string
Test Cases Input aaaabbbbcccc
Test Cases Expected Output Magic string
```




**Input:**
```
str //input String
```
**Logic:-**
Used while loop to test All conditions
```
 if(len(even)!=len(odd)):
        even=even.replace(even[int(len(even)/2)],"") #If the element is odd
                                                replace middle character with ""
    if(even==odd):
        print("Magic string")
    else:
        print("Not a magic string")
```
**OUTPUT:-**
```
Magic String
Not Magic String
```

```
A factory has ‘N’ employees working in it. Each employee has
an unique identification number (id). When they enter the
factory, their in_time is noted and when they leave the factory,
their out_time is noted. Both in_time and out_time of employees
can not be altered and in 24-hour format as hours:minutes. The
employees are paid hourly wages such that an employee gets Rs.
200/hour and Rs.3/minute if minutes>=30. Given the number of
employees, their id, in_time and out_time, develop an algorithm
and python code to print the id, in_time, out_time as strings and
wages for each employee with id as key and in_time, out_time
and wages as its value.
Note: use import pprint and pprint.pprint(variable) to print the
dictionary
Input format:
Enter the number of employees, N
Enter the id of 1st employee
Enter the in_time of 1st employee
Enter the out_time of 1st employee
…
Enter the id of nth employee
Enter the in_time of nth employee
Enter the out_time of nth employee
Output format:
print id as key and in_time, out_time, wages as values
```

**Input:-**
```
n //Number of Employees
id //Emp id
in_time //in_time of employee
out_time //out__time of employee
data // Dictionary for storing final outcome
```
**Logic**
```
for i in range(0,n):
  in_hr='' \\initialize variable
  in_min=''
  id=input('Enter the id of Employee:-')
  in_time=input('Enter the in_time of Employee:-')
  out_time=input('Enter the out_time of Employee:-')
  flag=0
  for i in in_time:
    if i==':':    //User may insert in 8:00 or 08:00 for getting appropriate i/p
      flag=1     //flag to difference between min & hr flag=0 for hr & 1 for min
    elif flag==0:
      in_hr+=i
    elif flag==1:
      in_min+=i
  in_hr=(int(in_hr)*60)+int(in_min)    //Convert in_hr into min
  out_hr=int(out_time[0:2])*60        //convert out_hr into min
  out_min=int(out_time[3:])       
  out_hr+=out_min
  hr=(out_hr-in_hr)//60              //convert to hour
  min=(out_hr-in_hr)%60              //convert to min
  if min>=30:                     //to calculate wage if min>30 based on hr&min
    wage=hr*200+min*3
  else:                           //calculate wage if min<30 based on hr only
    wage=hr*200
  lst[0]=in_time
  lst[1]=out_time
  lst[2]=wage                 // store all values in list lst
  t1=tuple(lst)              //store value of lst in tuple as tuple is immutable
  data[id]=t1                //insert data as key and value in dictionary data
  lst.clear                  //clear list for next iteration
  t1=tuple()                //to clear tuple
  s={k:data[k] for k in sorted(data)} //to sort element based on key in dict
```
**OUTPUT:-**
```
s //Sorted dictionary
```



````
'''
Test Cases Input
2
100
8:00
16:30
200
9:35
18:20
Test Cases Expected Output {'100': ('8:00', '16:30', 1690), '200': ('9:35', '18:20',
1735)}

'''
n=int(input("Enter the number of Employees:-"))
data={}
lst=['','','']
for i in range(0,n):
  in_hr=''
  in_min=''
  id=input('Enter the id of Employee:-')
  in_time=input('Enter the in_time of Employee:-')
  out_time=input('Enter the out_time of Employee:-')
  flag=0
  for i in in_time:
    if i==':':
      flag=1
    elif flag==0:
      in_hr+=i
    elif flag==1:
      in_min+=i
  in_hr=(int(in_hr)*60)+int(in_min)
  out_hr=int(out_time[0:2])*60
  out_min=int(out_time[3:])
  out_hr+=out_min
  hr=(out_hr-in_hr)//60
  min=(out_hr-in_hr)%60
  if min>=30:
    wage=hr*200+min*3
  else:
    wage=hr*200
  lst[0]=in_time
  lst[1]=out_time
  lst[2]=wage
  t1=tuple(lst)
  data[id]=t1
  lst.clear
  t1=tuple()
s={k:data[k] for k in sorted(data)}
print(s)
````
