A submarine dives ‘X’ feet for every ‘t’ minutes. Let the
temperature of the ocean surface water (at depth 0 meter) is
about T degree Celsius. Assume that the temperature decreases
as T/2, T/3, T/4 … so on for every 125 meters. Develop the
python code for displaying the depth in meters (whole number)
and temperature of depth of the ocean where the submarine is
after a day (with 2 decimal places). While computing
temperature, use any operator other than +, -, /, *.
Note: X, t and T are positive integers. 1 feet = 0.3048 meters
Input format:
Enter the depth in feet that the submarine dives
Enter the time in minutes
Enter the temperature of ocean surface water
Output format:
Depth in meters
Temperature of depth of the ocean where the submarine is

```
depth=int(input("Enter the Depth:-"))
time=int(input("Enter times in minutes:-"))
temprature=int(input("Enter the temprature:-"))
speed=(depth*0.3048)*(time**(-1))
met=speed*(24*60)
n=met*(125**(-1))
temp=temprature*(n**(-1))
print(met)
print(format(temp,'.2f'))
```

A lucky game is played by a player in which he has to feed two
values X1 and X2. The game computes the sum of digits of X1
and if it is greater than or equal to X2, then the player wins or
lose otherwise. Write python code to display the sum of digits of
X1 and the status as “Win” or “Lose”
Note: X1 and X2 are positive integers.
For example, if X1 is 1478 and X2 is 15, then the sum of digits
of X1 =1+4+7+8 = 20 > X2 and hence the player wins\
Input format:
Enter a number, X1
Enter another number, X2
Output format:
Sum of digits of X1
“Win” or “Lose”


```
x1=int(input("Enter No:-"))
x2=int(input("Enter No:-"))
result1=0
while x1 > 0:
    rem = x1 % 10
    result1 = result1 + rem
    x1 = int(x1/10)
result2=0
rem=0
while x2 > 0:
    rem = x2 % 10
    result2 = result2 + rem
    x2 = int(x2/10)
if result1>result2:
  print("The Number X1 Wins")
else:
  print("The Number X1 Lose")
```

A super number is a number whose reversed digits of right part
of the number are equal to the digits of left part of the number. If 
the number of digits of the given number is odd, then exclude
the middle number. Write python code to check whether the
given number is Super number or not.
For example, If the number is 12321, then reverse of right part is
12 which is equal to left part 12. Hence it is a super number. If
the number is 1234, then reverse of right part is 43 which is not
equal to left part 12. Hence, it is not a super number.
Input format:
Enter a number, X
Output format:
“Super number” or “Not a Super number”


```
x1=int(input("Enter No:-"))
rev = 0
x2=x1

while(x1 > 0): 
    a = x1 % 10
    rev = rev * 10 + a
    x1 = x1 // 10
if rev==x2:
  print("Super Number")
else:
  print("Not a Super Number")
```
