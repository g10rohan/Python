

```
Description
Automorphic numbers are numbers having "n" digits such that
the last n digits of the square of the number will be the number
itself. Write a Python code to check if the given number is
automorphic. Write a function to find the square of a given
number.. For example, 25 is a 2 digit automorphic number with
a square of 625 and 376 with its square 141376, is a 3 digit
automorphic number.
Input Format
First line contains the number to be checked
Output Format
Print Automorphic or Not automorphic
```



**Input:-**
```
n //Number to check
```
**Logic:-**


```
sqr = a**2 //calculate square of given number
  size= len(str(a))  //to calculate the Length
  l = sqr%10**size  // dividing sqr by 10 with power of size of number to get 
                      last digits  
  if l == a:       //Check whether the last digits are same as input number
    print("Automorphic Number")
  else:
    print("Not an Automorphic Number")
```
**Output:-**

```
 Automorphic Number or Not an Automorphic
```




````
'''
Test Cases Input 25
Test Cases Expected Output Automorphic
'''
n=int(input("Enter Number"))
def checkAutomorphic(a):
  sqr = a**2 
  size= len(str(a))
  l = sqr%10**size 
  if l == a:
    print("Automorphic Number")
  else:
    print("Not an Automorphic Number")
checkAutomorphic(n)
````

```
Given ‘n’ integers, write an algorithm and the subsequent
Python code to print all numbers that are isomorphic to the first
number. Two numbers with non-distinct digits (numbers in
which digits get repeated) are said to be isomorphic if they have
the same number of digits in it and the sets of positions having
the same digits are equal. For example:
 12321 is isomorphic to 83538 .
Both the numbers are of length five. In 12321, positions 2 and 4
have the same digit 2, Positions 1 and 5 have same digit 1.
Hence the set of positions having the same digit are
{{1,5},{2,4}. Similarly , for the number 83538, the set of
positions having the same digit is {{1,5}, {2,4}}.
 1232 is not isomorphic to 2342 because set of places
having same digits for 1232 is {{2, 4}} and for 2342 is
{{1, 4}} (digits are numbered from left to right starting 
from 1).
 12 is not isomorphic to 10.
Write a function to check whether two numbers are
isomorphic. If none of the numbers are isomorphic then print
‘No isomorphic’.
Input Format
First line contains the number of elements, n
Next ‘n’ line contains the numbers
Output Format
Print first number in the first line
Next few lines contain the numbers that are isomorphic to first
number.
If the none of the numbers are not isomorphic, then Print “No
isomorphic”
```

````
'''
Test Cases Input
5
12321
83538
12312
2345
45654
Test Cases Expected Output
12321
83538
45654
'''
n = int(input())
def is_isomorphic(x):
    l= set()
    for a in x:
        if x.count(a)>1:
            p= list()
            for b in range(len(x)):
                if a==x[b]:
                    p.append(b+1)
            if len(p)>0:
                l.add(tuple(p))
    if len(l)>1:
        return(l)
    else:
        return('n')
fno = input().rstrip()
f = 0
l=[]
for a in range(1,n):
    x = input().rstrip()
    if len(fno)==len(x) and is_isomorphic(fno)!='n' and is_isomorphic(x)!='n' and is_isomorphic(fno)==is_isomorphic(x):
        l.append(x)
        f=1
if f == 0:
    print('No isomorphic')

else:
    print(fno)
    for a in l:
        print(a)
````
